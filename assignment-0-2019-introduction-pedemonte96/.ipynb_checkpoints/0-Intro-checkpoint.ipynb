## Consolidació dels coneixements i ajuda

Per **auto-evaluar** els coneixements apressos, es disposa d'una plataforma amb la qual es pot avaluar la feina feta. Abans d'entrar en detalls de com fer-ho, cal tenir present els següents punts:

* La nota de la plataforma d'autoavaluació **no** reflecteix de forma directa la d'avaluació de la pràctica. Treure un 10 en l'autoavaluació **no** implica treure un 10 en la pràctica. Hi ha criteris que l'autoavaluació no pot determinar, d'entre altres:
  * Estil i netedat del codi
  * Eficiència i optimització
  * Ús de les eines disponibles
  * ...
* El sistema no és a prova de tot. Cal seguir exactament les especificacions donades per tal que l'avaluació sigui correcta. Si la documentació diu que la funció ha de retornar un enter, retornar qualsevol altra cosa (com ara un flotant) pot implicar un error en la funció
* L'ús de l'autoavaluador està limitat per dia i grup. Normalment es permeten 3 intents diaris per cada grup.
* Els consells que dóna l'autoavaluador són genèrics. Si bé haurien de servir d'ajuda, no tenen perquè reflectir exactament el problema del vostre codi
* L'autoavaluador està subjecte a canvis i pot presentar problemes. Si marqués error en una funció on s'està convençut que no n'hi ha, caldrà contactar amb el professorat.
* La disponibilitat de l'autoavaluador **NO** està assegurada. És (més) que probable que en moments d'alta demanda (per exemple el dia abans d'una entrega, examen, etc.) aquest funcioni molt lent (o directament no funcioni).

### Com realitzar un test

Per defecte els commits a Github **no** realitzen testos, en cas contrari seria molt fàcil superar el límit diari de forma no intencionada. Recordeu: els commits han de ser concisos i, dins del possible, centrats en un canvi particular (vs grans i genèrics).

Per fer el test, el primer serà tenir un un repositori clonat:
```bash
git clone urlRespository
```

i haver fet un commit (i, opcionalment, pujat a Github):

```bash
git add .
git commit -m "Breu explicació del contingut del commit"
git push origin master
```

Fins aquí, res fora del normal. Ara, per executar el test caldrà fer:
```bash
git push origin master:test
```

L'execució i resultats dels tests els podreu comprovar des de:
<div style="text-align: center;"><a href="http://grade-me.education" target="_blank">http://grade-me.education</a></div>

Els testos comproven:

* La correcta execució del codi
* El resultat de les funcions
* L'utilització de codi "correcta", com ara no incloure bucles, no cridar certes funcions, etc.

No tots els testos són públics, alguns d'ells solament són visibles pels professors. D'igual forma, en cas de que una funció no passi el test, no sempre podràs veure que ha fallat.

### Troubleshooting

**Mai canvieu el nom del notebook**

**Mai canvieu el nom de les funcions**

**Mai deixeu codi de prova fora de blocs `if __name__ == '__main__'`**

**Mai feu servir llibreries/codi extern**. _A no ser que la pràctica ho indiqui_

**Mai assumiu que els tests es fan en un ordre particular**. 

Sobretot, cobra rellevància si feu coses rares com mantenir un estat (que mai demanarem en les pràctiques).

```python

estat = 1

def fun1(param):
    global estat
    if estat == 1:
        estat = 2
        
    return param * estat
```
Fallarà al comprovar
```python
fun1(2) == 2
fun1(3) == 3
``` 

Si fos necessari fer una sèrie de crides (perquè així ho digui l'enunciat), els testos també ho faran. En aquest cas es fa referència a comportament no esperat.

**He fet els tests i em diu que hi ha un error en importar el meu codi**

El més probable és que tinguis una prova fora d'un bloc `if __name__ == '__main__'` i aquesta depengui d'una altra prova anterior que sí que està ben definida. En importar s'executa el codi sencer excepte aquests blocs, fent saltar un error per fer servir variables (i/o funcions) no definides.

Revisa el teu codi i assegurat que executi bé.

**Em diu que tot està malament i no crec que sigui així**

Revisa el `Log`, assegura't que no indiqui un error en importar. En cas d'error d'importació, tots els testos fallaran.

Si no ha fallat en importar, revisa que segueixis les especificacions dels tipus dels paràmetres i retorns.

Si estàs convençut que està tot bé, contacta amb el professorat.
